# **Feature Engineering**

```python
# Create a new feature that counts the total number of mental health issues each student has
df['Total Mental Health Issues'] = df['Have you been feeling more negative than usual?'] + df['Have you been sleeping worse than usual?'] + df['Over the last 3 months, Did you feel alone or lonely?']+ df['Are You having any extreme emotions or mood swings? (breakdowns, or panic attacks)'] + df['Do you often struggle to maintain a healthy balance between yourself, your career, physical activity, and those you care about?'] + df['Do you have/had  backlog?']+df['Do you have a pending assignment?']+df['Are you having trouble with your attendence?']+df['Are you able to meet the deadline for assignments?']
```

```python
# Convert 'Your current year of Study' to a numerical value
df['Study Year'] = df['Your current year of Study'].map({
    'year 1': 1,
    'year 2': 2,
    'year 3': 3,
    'year 4': 4,
    'year 5': 5,
})
```

```python
df['CGPA Midpoint'] = df['What is your CGPA?'].map({
    '3.00 - 3.49': 3.25,
    '3.50 - 3.99': 3.75,
    '2.50 - 2.99': 2.75,
    '2.00 - 2.49': 2.25,
    '1.50 - 1.99': 1.75,
    '1.00 - 1.49': 1.25,
    '0.50 - 0.99': 0.75,
    '0.00 - 0.49': 0.25
})

df = df.drop('What is your CGPA?', axis=1)
```

# **Model Selection**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Define feature matrix X and target vector y
X = df.drop('Total Mental Health Issues', axis=1)
y = df['Total Mental Health Issues']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}
```

# **Model Training**

```python
num_cols = ['Age']
cat_cols = ['Choose your gender', 'What is your course?', 'Your current year of Study', 'Are you in relationship ?', 'Have you been feeling more negative than usual?', 'Have you been sleeping worse than usual?', 'Over the last 3 months, Did you feel alone or lonely?', 'Are You having any extreme emotions or mood swings? (breakdowns, or panic attacks)', 'Do you often struggle to maintain a healthy balance between yourself, your career, physical activity, and those you care about?']
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the median for numerical columns
num_imputer = SimpleImputer(strategy='median')
X_train[num_cols] = num_imputer.fit_transform(X_train[num_cols])
X_test[num_cols] = num_imputer.transform(X_test[num_cols])

# Impute missing values with the most frequent category for categorical columns
cat_imputer = SimpleImputer(strategy='most_frequent')
X_train[cat_cols] = cat_imputer.fit_transform(X_train[cat_cols])
X_test[cat_cols] = cat_imputer.transform(X_test[cat_cols])
```

```python
# Impute missing values in the 'CGPA Midpoint' column with the median value
median_value = X_train['CGPA Midpoint'].median()
X_train['CGPA Midpoint'].fillna(median_value, inplace=True)
X_test['CGPA Midpoint'].fillna(median_value, inplace=True)
```

```python
# Drop the 'Study Year' column from the training and test sets
X_train = X_train.drop('Study Year', axis=1)
X_test = X_test.drop('Study Year', axis=1)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

# **Model Evaluation**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Define a function to calculate and print the evaluation metrics
def print_evaluation_metrics(y_test, y_pred, average='micro'):
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"Precision: {precision_score(y_test, y_pred, average=average)}")
    print(f"Recall: {recall_score(y_test, y_pred, average=average)}")
    print(f"F1 Score: {f1_score(y_test, y_pred, average=average)}")
```

```python
# Train the models
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)
```

```python
# Predict the target variable for the test set using each model
for model_name, model in models.items():
    print(f"Model: {model_name}")
    y_pred = model.predict(X_test)
    print_evaluation_metrics(y_test, y_pred)
    print("\n")
```
